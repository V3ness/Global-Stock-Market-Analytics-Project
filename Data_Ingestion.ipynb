# Date Ingestion (Bronze Layer)

```python
# Load csv file from volume into dataframe
df_raw = spark.read.format("csv") \
            .option("header", "true") \
            .option("inferschema", "true") \
            .load("/Volumes/gsma_workspace/global_stocks_schema/global_stocks/Index_Levels.csv")

# Display the dataframe
display(df_raw)
```

```python
# Check schema
df_raw.printSchema()
```

```python
# Validate row counts
print(f"Row count: {df_raw.count()}")
```

```python
# Save dataframe as Delta format and creates a bronze table
df_raw.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("gsma_workspace.global_stocks_schema.bronze_global_stock_data")
```

```python
%sql
-- Validate table created
SELECT COUNT(*) FROM gsma_workspace.global_stocks_schema.bronze_global_stock_data;
```
